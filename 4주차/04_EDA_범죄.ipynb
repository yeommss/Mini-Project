```python
# 두번쨰 데이터 로드
raw2 = pd.read_csv('data/범죄_발생부터_검거까지의_기간_2011_2020.csv', encoding='cp949')
raw2.shape
```

```python
# 컬럼명 변경
raw = raw.rename(columns={
    "죄종별(1)": "범죄종류", "죄종별(2)": "범죄명",
    "재범종류별(1)": "재범종류", "재범기간별(1)": "기간",})

# 바뀐 컬럼명 확인
raw.columns
```

```python
raw2 = raw2.rename(columns={
    "죄종별(1)": "범죄종류", "죄종별(2)": "범죄명",
    "범죄발생검거기간별(1)": "검거기간", })

raw2.columns
```

<center><h2><b>Tidy Data 만들기</b></h2></center>


```python
# Tidy Data 만들기!
df = pd.melt(raw, id_vars=['범죄종류', '범죄명', '재범종류', '기간'],
            var_name='연도', value_name='재범발생횟수')
df.head(10)
```

```python
# Tidy Data 만들기!
df2 = pd.melt(raw2,
            id_vars=['범죄종류', '범죄명', '검거기간'],
            var_name='연도', value_name='범죄자수')
df2.head(10)
```

```python
# 요약 확인
df.info()
```

```python
# 타입 확인
df.dtypes, df2.dtypes
```

```python
# 연도, 발생횟수 수치형 데이터로 변경
# df['연도'] = df['연도'].astype(int)
# df['연도'].dtypes

df2['연도'] = df2['연도'].astype(int)
df2['연도'].dtypes
```

```python
# 연도, 발생횟수 수치형 데이터로 변경
# 발생횟수 같은 경우는 astype 을 이용하면 중간에 string 오류가 난다.
# downcast 하면 타입 변경 가능 -> 근데 결측치가 있어서 float
df['재범발생횟수'] = pd.to_numeric(df['재범발생횟수'], errors='coerce', downcast='integer')
df['재범발생횟수'].dtypes

```

```python
df2['범죄자수'] = pd.to_numeric(df2['범죄자수'], errors='coerce', downcast='integer')
df2['범죄자수'].dtypes
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
# 결측치 변경
df['재범발생횟수'] = df['재범발생횟수'].fillna(0).astype(int)
```

```python
# 결측치 변경 확인
df.isnull().sum()
```

```python
# 결측치 변경
df2['범죄자수'] = df2['범죄자수'].fillna(0).astype(int)
```

```python
df2.isnull().sum()
```

```python
# 고유값 확인하기
for col in df.columns:
    print(col, ':', end = ' ')
    print(df[col].unique(), end = '\n\n')
```

```python
# 소계 하나뿐인지 확인
df2.groupby(by=['범죄종류', '범죄명']).count()
```

```python
# 소계 들어가는 행 삭제
drops_col = ['강력범죄', '지능범죄', '폭력범죄', '풍속범죄']
drops = ((df['범죄명'].str.contains('계')) & (df['범죄종류'] == drops_col[0])) |\
((df['범죄명'].str.contains('계')) & (df['범죄종류'] == drops_col[1])) |\
((df['범죄명'].str.contains('계')) & (df['범죄종류'] == drops_col[2])) |\
((df['범죄명'].str.contains('계')) & (df['범죄종류'] == drops_col[3])) |\
    df['범죄종류'].str.contains('계') | df['기간'].str.contains('계')
df = df[~drops]
```

```python
df.tail()
```

```python
# drops_col = ['강력범죄', '지능범죄', '폭력범죄', '풍속범죄']
drops = ((df2['범죄명'].str.contains('계')) & (df2['범죄종류'] == drops_col[0])) |\
((df2['범죄명'].str.contains('계')) & (df2['범죄종류'] == drops_col[1])) |\
((df2['범죄명'].str.contains('계')) & (df2['범죄종류'] == drops_col[2])) |\
((df2['범죄명'].str.contains('계')) & (df2['범죄종류'] == drops_col[3])) |\
    df2['범죄종류'].str.contains('계') | df2['검거기간'].str.contains('계')
df2 = df2[~drops]
```

```python
df2.tail()
```

```python
# 소계 겹치지 않게 범죄명 변경
for col in df['범죄종류'].unique().tolist():
    if col not in drops_col:
        df.loc[df['범죄종류'] == col, '범죄명'] = col
```

```python
for col in df2['범죄종류'].unique().tolist():
    if col not in drops_col:
        df2.loc[df2['범죄종류'] == col, '범죄명'] = col
```

```python
# 제대로 변경 되었는지 확인
df.groupby(by=['범죄종류', '범죄명']).count()
# df2.groupby(by=['범죄종류', '범죄명']).count()
```

```python
# 인덱스 재정의
df = df.reset_index(drop=True)
df.head()
```

```python
df2 = df2.reset_index(drop=True)
df2.head()
```

```python
# 컬럼 재정렬
cols = ['범죄명', '연도', '기간', '재범종류', '범죄종류', '재범발생횟수', ]
df = df[cols]
```

```python
# 컬럼 재정렬
cols = ['범죄명', '연도', '검거기간', '범죄종류', '범죄자수', ]
df2 = df2[cols]
```

```python
# 데이터 저장
df.to_csv('output/04_재범종류_및_기간_2011_2020.csv', index=False)
df2.to_csv('output/04_검거까지_기간_2011_2020.csv', index=False)
```
