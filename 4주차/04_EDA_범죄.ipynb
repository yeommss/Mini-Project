```python
# 데이터 로드
df_raw = pd.read_csv('data/재범자_재범종류_및_기간_2011_2020.csv', encoding='cp949')
df_raw.shape
```

```python
# 앞에
df_raw.head()
```

```python
# 컬럼명 확인
df_raw.columns
```

```python
# 컬럼명 변경
df_raw = df_raw.rename(columns={
    "죄종별(1)": "범죄종류", "죄종별(2)": "범죄명",
    "재범종류별(1)": "재범종류", "재범기간별(1)": "기간",})

# 바뀐 컬럼명 확인
df_raw.columns
```

---
----

<center><h2><b>Tidy Data 만들기</b></h2></center>


```python
# Tidy Data 만들기!
df = pd.melt(df_raw, id_vars=['범죄종류', '범죄명', '재범종류', '기간'], var_name='연도', value_name='재범발생횟수')
df.head(10)
```

```python
# 요약 확인
df.info()
```

```python
# 타입 확인
df.dtypes
```

```python
# 연도, 발생횟수 수치형 데이터로 변경
df['연도'] = df['연도'].astype(int)
df['연도'].dtypes
```

```python
# 연도, 발생횟수 수치형 데이터로 변경
# 발생횟수 같은 경우는 astype 을 이용하면 중간에 string 오류가 난다.
# downcast 하면 타입 변경 가능 -> 근데 결측치가 있어서 float
df['재범발생횟수'] = pd.to_numeric(df['재범발생횟수'], errors='coerce', downcast='integer')
df['재범발생횟수'].dtypes
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 결측치 변경
df['재범발생횟수'] = df['재범발생횟수'].fillna(0).astype(int)
```

```python
# 결측치 변경 확인
df.isnull().sum()
```

```python
# 고유값 확인하기
for col in df.columns:
    print(col, ':', end = ' ')
    print(df[col].unique(), end = '\n\n')
```

```python
# '소계' 들어가는 행 삭제
drops = df['범죄종류'].str.contains('계') | df['범죄명'].str.contains('계') \
    | df['재범종류'].str.contains('계') | df['기간'].str.contains('계')
df = df[~drops]
df.head()
```

```python
# 인덱스 재정의
df = df.reset_index(drop=True)
df.head()
```

```python
# 컬럼 재정렬
cols = ['범죄명', '연도', '기간', '재범종류', '범죄종류', '재범발생횟수', ]
df = df[cols]
```

```python
# 데이터 저장
df.to_csv('output/04_재범종류_및_기간_2011_2020.csv', index=False)
```
