```python
import pandas as pd
import numpy as np
```

```python
air_raw = pd.read_csv('data/air/air.csv')
air_raw.head()
```

```python
air_19 = air_raw[air_raw['날짜'].astype('string').str.startswith('2019')]
air_20 = air_raw[air_raw['날짜'].astype('string').str.startswith('2020')]
```

```python
air_20.head()
```

```python
air_19 = air_19.rename(columns={
    '지역': '시도명',
    '종류': '항목',
    '날짜': '월',
    '수치': '2019'
})
air_19.head()
```

```python
air_20 = air_20.rename(columns={
    '지역': '시도명',
    '종류': '항목',
    '날짜': '월',
    '수치': '2020'
})
air_20.head()
```

```python
# 월 데이터를 숫자만 반환하도록 설정한다.
def find_month(x):
    # [2020.01] 이런식으로 되어있는거 자르고 0 있으면 없애기
    return x.split('.')[-1]

air_19['월'] = air_19['월'].astype('string').map(find_month)
air_20['월'] = air_20['월'].astype('string').map(find_month)
air_20.head()
```

```python
air_20.loc[:, '2020']
```

```python
air_19.shape, air_20.shape
```

```python
air_19.dtypes
```

```python
air_1920 = pd.merge(air_19, air_20)
air_1920.to_csv('output/air/air_1920.csv', index=False)
```

#### 19년 mean, median, max, min

```python
air_19_mean = air_19.groupby(by=['시도명', '항목']).mean()
air_19_mean  = air_19_mean.rename(columns={'2019': '19년_대기_평균'})
air_19_mean.head()
```

```python
air_19_median = air_19.groupby(by=['시도명', '항목']).median()
air_19_median = air_19_median.rename(columns={'2019': '19년_대기_중앙치'})
air_19_median.head()
```

```python
air_19_max = air_19.groupby(by=['시도명', '항목']).max()
air_19_max = air_19_max.rename(columns={'2019': '19년_대기_최댓값'})
air_19_max = air_19_max.drop(columns='월')
air_19_max.head()
```

```python
air_19_min = air_19.groupby(by=['시도명', '항목']).min()
air_19_min = air_19_min.rename(columns={'2019': '19년_대기_최솟값'})
air_19_min = air_19_min.drop(columns='월')
air_19_min.head()
```

```python
air_19 = pd.merge(air_19_mean, air_19_median, left_index=True, right_index=True)
air_19 = pd.merge(air_19, air_19_max, left_index=True, right_index=True)
air_19 = pd.merge(air_19, air_19_min, left_index=True, right_index=True)
air_19
```

#### 20년 mean, median, max, min

```python
air_20_mean = air_20.groupby(by=['시도명', '항목']).mean()
air_20_mean  = air_20_mean.rename(columns={'2020': '20년_대기_평균'})
air_20_mean.head()
```

```python
air_20_median = air_20.groupby(by=['시도명', '항목']).median()
air_20_median = air_20_median.rename(columns={'2020': '20년_대기_중앙치'})
air_20_median.head()
```

```python
air_20_max = air_20.groupby(by=['시도명', '항목']).max()
air_20_max = air_20_max.rename(columns={'2020': '20년_대기_최댓값'})
air_20_max = air_20_max.drop(columns='월')
air_20_max.head()
```

```python
air_20_min = air_20.groupby(by=['시도명', '항목']).min()
air_20_min = air_20_min.rename(columns={'2020': '20년_대기_최솟값'})
air_20_min = air_20_min.drop(columns='월')
air_20_min.head()
```

```python
air_20 = pd.merge(air_20_mean, air_20_median, left_index=True, right_index=True)
air_20 = pd.merge(air_20, air_20_max, left_index=True, right_index=True)
air_20 = pd.merge(air_20, air_20_min, left_index=True, right_index=True)
air_20
```

```python
air_1920 = pd.merge(air_19, air_20, left_index=True, right_index=True)
air_1920.to_csv('output/air/air_stat_1920.csv')
```
